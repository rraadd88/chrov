# ðŸ§¬Visualization of genome, set of chromosomes, a chromosome and an arm

```python
import pandas as pd
import matplotlib.pyplot as plt

from chrov.viz.chrom import plot_chroms
```

```python
## input parameters
chromosome='1'
arm='p'
test=False
input_dir_path='../modules/inputs'
output_dir_path='outputs'
```

```python
## set the current directory to 'examples'
import os
if len(set(['chrov','examples','tests']) - set(os.listdir('.')))==0:
    ## change the current directory from 'chrov' to 'examples'
    os.chdir('./examples')
```

```python
## process inputs
from os import makedirs
makedirs(output_dir_path,exist_ok=True)
```

```python
# cytobands to plot the chromosome
cytobands=pd.read_table(f'{input_dir_path}/cytobands.tsv',index_col=[0])
cytobands.head(1)
```

## Genome

### linear layout

```python
ax=plot_chroms(
    data=cytobands,
    arc=False,
    show_labels=True,
    label_y=0.05,
    test=test,
    figsize=[20,1],
)
_=ax.set(ylim=[-0.2,0.2])
```

### polar layout

```python
ax=plot_chroms(
    data=cytobands,
    # arc=True, default for plot_chroms
    test=test,
    figsize=[5,5],
)
_=ax.set(ylim=[-1,0.2])
```

```python
ax=plot_chroms(
    data=cytobands,
    # arc=True, default for plot_chroms
    pi_span=2,
    test=test,
    figsize=[4,4],
)
_=ax.set(ylim=[-1,0.2])
```

```python
# other than default settings 
fig,axs=plt.subplots(2,2,subplot_kw=dict(projection='polar'))
for ax,pi_span in zip(axs.ravel(),[2,1.5,1,0.5]):
    ax=plot_chroms(
        data=cytobands,
        pi_span=pi_span,
        pi_start=2-pi_span,
        lw=5,
        test=test,
        show_labels=False,
        ax=ax,
    )
    _=ax.set(title=f"pi_span={pi_span}, pi_start={2-pi_span}",ylim=[-1,0.2])
plt.savefig(f"{output_dir_path}/pi_starts_spans.png",bbox_inches='tight')
```

## Chromosomes

### linear layout

```python
ax=plot_chroms(
    data=cytobands,
    arc=False,
    chromosomes=[1,2,3,4],
    show_labels=True,
    figsize=[20,0.5],
    test=test,
    )
```

### polar layout

```python
fig=plt.figure()
ax=plot_chroms(
    data=cytobands,
    chromosomes=[1,2,3,4],
    show_labels=True,
    figsize=[5,5],
    test=test,
)
_=ax.set(ylim=[-1,0.2])
```

```python
fig=plt.figure()
ax=plot_chroms(
    data=cytobands,
    chromosomes=[1,2,3,4],
    show_labels=True,
    pi_span=2,
    figsize=[5,5],    
    test=test,
)
_=ax.set(ylim=[-1,0.2])
```

## Chromosome

### linear layout

```python
ax=plot_chroms(
    data=cytobands.query(expr=f"`chromosome` == '{chromosome}'"),
    arc=False,
    show_labels=True,
    figsize=[4,0.2],
    test=test,
    )
```

### polar layout

```python
fig=plt.figure()
ax=plot_chroms(
    data=cytobands.query(expr=f"`chromosome` == '{chromosome}'"),
    show_labels=True,
    test=test,
)
_=ax.set(ylim=[-1,0.2])
```

## Chromosome arm

### linear layout

```python
ax=plot_chroms(
    data=cytobands.query(expr=f"`chromosome` == '{chromosome}' & `arm` == '{arm}'"),
    arc=False,
    show_labels=True,
    figsize=[4,0.2],
    test=test,
    )
```

### polar layout

```python
fig=plt.figure()
ax=plot_chroms(
    data=cytobands.query(expr=f"`chromosome` == '{chromosome}' & `arm` == '{arm}'"),
    show_labels=True,
    test=test,
)
_=ax.set(ylim=[-1,0.2])
```
